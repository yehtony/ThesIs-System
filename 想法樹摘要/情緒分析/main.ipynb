```python
from transformers import BertForSequenceClassification
from transformers import BertTokenizer, pipeline
import torch

tokenizer=BertTokenizer.from_pretrained('IDEA-CCNL/Erlangshen-Roberta-330M-Sentiment')
model=BertForSequenceClassification.from_pretrained('IDEA-CCNL/Erlangshen-Roberta-330M-Sentiment')

sentiment_analyzer= pipeline("text-classification", model=model, tokenizer=tokenizer)

# output=model(torch.tensor([tokenizer.encode(text)]))
# print(torch.nn.functional.softmax(output.logits,dim=-1))
```

```python
import pandas as pd
import numpy as np

# 使用相对路径加载CSV文件
csv_file_path = "../想法樹假資料處理/ideatree_00.csv"

# 使用pandas加载CSV文件到DataFrame
df = pd.read_csv(csv_file_path)

# 打印DataFrame以查看数据
df
```

```python
df['sentiment'] = df['節點_內容'].apply(sentiment_analyzer).apply(lambda x: 'Negative' if x[0]['label'] == 'Negative'and x[0]['score'] > 0.8 else 'Positive')
```
