```python
import pandas as pd
import numpy as np

# 使用相对路径加载CSV文件
csv_file_path = "../想法樹假資料處理/ideatree_00.csv"

# 使用pandas加载CSV文件到DataFrame
df = pd.read_csv(csv_file_path)

# 打印DataFrame以查看数据
df
```

```python
# 生成目標節點的 DataFrame
target_df = df[["節點_id", "節點_類型"]].rename(columns={"節點_id": "目標節點_id", "節點_類型": "目標節點_類型"})

# 合併原始 DataFrame 和目標節點 DataFrame
merged_df = pd.merge(df, target_df, left_on="目標節點_id", right_on="目標節點_id", how="left")

# 顯示結果
print(merged_df[["節點_類型", "目標節點_類型"]])

df_main = merged_df
```

```python
transition_table = pd.pivot_table(df_main, index='節點_類型', columns='目標節點_類型', aggfunc='size', fill_value=0)
```

```python
# 计算事件的总数
total_events = len(df_main)

# 创建一个空的Z-score表格
z_score_table = pd.DataFrame(
    index=transition_table.index, columns=transition_table.columns
)
# 计算Z-score并填充表格
for from_value in transition_table.index:
    for to_value in transition_table.columns:
        # 事件 g 到事件 t 的次數
        f_g_t = transition_table.at[from_value, to_value]
        # print(f_g_t)
        # 事件 g 的總數
        f_g = transition_table.loc[from_value].sum()
        # 發生事件 g 的機率
        p_g = f_g / total_events
        # 發生事件 t 的機率
        p_t = transition_table[to_value].sum() / total_events
        # 计算Z-score
        z_score = (f_g_t - f_g * p_t) / np.sqrt(f_g * p_t * (1 - p_t) * (1 - p_g))
        # 将Z-score填充到表格
        z_score_table.at[from_value, to_value] = z_score

# 显示Z-score表格
z_score_table
```
