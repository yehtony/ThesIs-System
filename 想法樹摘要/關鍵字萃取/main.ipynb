```python
import jieba
from sklearn.feature_extraction.text import TfidfVectorizer

data["merged_notes"] = data["source_note"].fillna('') + ' ' + data["target_note"].fillna('')
display(data["merged_notes"])

# 使用TF-IDF算法提取关键词的函数
def extract_keywords_tfidf(text):
    if text is None or not text.strip():
        return ""
    words = list(jieba.cut(text))
    text = ' '.join(words)

    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform([text])
    feature_names = tfidf_vectorizer.get_feature_names_out()

    sorted_items = [(feature_names[i], score) for i, score in enumerate(tfidf_matrix.toarray()[0])]
    sorted_items = sorted(sorted_items, key=lambda x: x[1], reverse=True)
    keywords_tfidf = [item[0] for item in sorted_items[:5]]
    return ' '.join(keywords_tfidf)

# 应用TF-IDF算法提取关键词到整个文本
all_notes_text = ' '.join(data["merged_notes"].tolist())
all_notes_keywords = extract_keywords_tfidf(all_notes_text)

# 打印结果
print("TF-IDF关键词:", all_notes_keywords)
```

```python
from ckiptagger import data_utils, WS
from textrank4zh import TextRank4Keyword

# 下载 CKIP 模型数据
data_utils.download_data_gdown("./")

# 初始化 CKIP 模型
ws = WS("./data")

# 定义文本
text = "太陽光會什麼可以發電？"

# 对文本进行分词
word_list = ws([text])

# 创建 TextRank4Keyword 实例
tr4w = TextRank4Keyword()

# 将分词结果传递给 TextRank4Keyword
tr4w.analyze(text, lower=True, window=2)

# 获取关键词
keywords = tr4w.get_keywords(3)  # 获取前5个关键词

# 打印关键词
for keyword in keywords:
    print(keyword.word, keyword.weight)
```
